```python
# Tumor Growth Modeling using Logistic Equation

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score, mean_squared_error

# Logistic growth model
def logistic_model(t, r, K, N0):
    return (K * N0 * np.exp(r*t)) / (K + N0*(np.exp(r*t) - 1))

# Time points (days)
t = np.linspace(0, 20, 20)

# Simulated tumor volume data (with noise)
true_values = logistic_model(t, 0.35, 1000, 10)
observed_data = true_values + np.random.normal(0, 20, 20)

# Curve fitting
params, covariance = curve_fit(logistic_model, t, observed_data, p0=[0.3, 1000, 10])

r_est, K_est, N0_est = params

# Predictions
predicted = logistic_model(t, r_est, K_est, N0_est)

# Model evaluation
r2 = r2_score(observed_data, predicted)
rmse = np.sqrt(mean_squared_error(observed_data, predicted))

print("Estimated Growth Rate (r):", r_est)
print("Estimated Carrying Capacity (K):", K_est)
print("R2 Score:", r2)
print("RMSE:", rmse)

# Plot
plt.scatter(t, observed_data, label="Observed Data")
plt.plot(t, predicted, label="Logistic Model Fit")
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume")
plt.legend()
plt.title("Tumor Growth Modeling using Logistic Equation")
plt.show()
```

The logistic model captures tumor growth under resource limitations. The intrinsic growth rate (r) represents the proliferation rate of cancer cells, while the carrying capacity (K) reflects environmental constraints such as nutrient limitation and space restriction. The high RÂ² value indicates that the logistic model effectively explains the observed tumor growth dynamics. This demonstrates how mathematical modeling can be used to understand biological systems quantitatively.
